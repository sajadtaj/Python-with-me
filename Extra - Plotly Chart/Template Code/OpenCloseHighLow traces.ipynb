# import Pack

```python
import pandas as pd
import plotly.graph_objects as graph
```

# Call Data

```python
candle = pd.read_pickle(r'D:\Plotly Chart\Template Code\Data\Candle Data\FEOLAD_CL_20200201_20220725.pickle')
```

```python
candle
```

# Create Candle

```python
# Step 1
fig = graph.Figure()

# Step 2
fig.add_traces(graph.Ohlc(x=candle['Date'],
    open=candle['priceFirst'],
    close=candle['pClosing'],
    high=candle['priceMax'],
    low=candle['priceMin'],)
)
# Step 3
fig.show()
```

## حذف رنج زمان

```python
fig.update(layout_xaxis_rangeslider_visible=False)
```

## Title & Yaxis

```python
fig.update_layout(
    title='The Great Recession',
    yaxis_title='AAPL Stock',
)
```

## اقزودن خط در تاریخ مشخص 

```python
fig.update_layout(
    shapes = [dict(
        x0='2021-12-09', x1='2021-12-09', y0=0, y1=1, xref='x', yref='paper',
        line_width=2)],
    annotations=[dict(
        x='2021-12-09', y=0.05, xref='x', yref='paper',
        showarrow=False, xanchor='left', text='Increase Period Begins')]
)
```

## افزودن مربع در یک تاریخ

```python
fig.update_layout(
    shapes = [dict(
        x0='2021-10-09', x1='2021-12-09', y0=0.2, y1=0.8, xref='x', yref='paper',
        line_width=2)],
    annotations=[dict(
        x='2021-12-09', y=0.05, xref='x', yref='paper',
        showarrow=False, xanchor='left', text='Increase Period Begins')]
)
```

## تغیر رنگ کندل ها

```python
# Step 1 
fig2 = graph.Figure()

# Step 2
fig2.add_traces(graph.Ohlc(x=candle['Date'],
    open=candle['priceFirst'],
    close=candle['pClosing'],
    high=candle['priceMax'],
    low=candle['priceMin'],
    increasing_line_color= 'cyan', 
    decreasing_line_color= 'gray',
    line_dash='solid')
   
)

# Step 3
fig2.show()
```
