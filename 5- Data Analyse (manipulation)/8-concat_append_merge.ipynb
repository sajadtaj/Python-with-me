```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8],'C':['c','c1','c2','c3']})
df2= pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8],'D':['d','d1','d2','d3']})
df3= pd.DataFrame({'A':[1,8,3,6],'B':[5,6,7,8],'E':['e','e1','e2','e3']})
```

```python
df1
```

```python
df2
```

## Append

```python
df_1and2 = df1.append(df2)
```

```python
df_1and2
```

```python
df_1and2and3 =df1.append([df2,df3])
df_1and2and3
```

## Merge (Join)

```python
df_merge_1and3= pd.merge(df1,df3,how='left')
df_merge_1and3
```

```python
df_merge_1and2= pd.merge(df1,df2,how='left')
df_merge_1and2
```

```python
df_merge_1and3= pd.merge(df1,df3,how='right',on='A')
df_merge_1and3
```

```python
df_merge_1and3= pd.merge(df1,df3,how='inner',on='A')
df_merge_1and3
```

```python
df_merge_1and3= pd.merge(df1,df3,how='left',on=['A','B'])
df_merge_1and3
```

## Concat
+   axis=1 --> Vertical
+   axis=0 --> Hor

```python
df_concat_1and2 = pd.concat([df1,df2],axis=1)
df_concat_1and2
```

```python
df_concat_1and2 = pd.concat([df1,df2],axis=0)
df_concat_1and2
```
