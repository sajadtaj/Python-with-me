```python
import pandas as pd
import numpy as np
from numpy import NaN
import plotly.express as px 
df1 = pd.DataFrame({'A':[1,2,NaN,4,NaN,15,45,85,45,45,78,NaN,1,25],
                    'B':[5,6,7,8,13,12,45,NaN,27,8,96,58,45,NaN],
                    'C':['c1','c1','c3',NaN,'c3','c3','b1','v1','v1',NaN,'d2','d2','d2','d1'],
                    'D':[NaN,1,0,1,0,1,0,0,NaN,1,1,0,1,0]})
```

### نمایش مقادیر از دست رفته

```python
df1.isna()
```

### تعداد مقادیر از دست رفته

```python
df1.isna().sum()
```

# اگر مقادبر از دست رفته با نان مشخص نشده باشند
+  با حرف یاعددی دیگر مشخص شده باشند 

```python
# مقادبر آنها را با نان عوض میکنیم
df1[df1=='?'] = np.nan
```

## حذف مقادبر از دست رفته
+   حذف رکوردی که نال دارد (سطر)

```python
new_df_0 =df1.dropna(axis=0)
new_df_0
```

```python
## حذف ستونهای که نتل دارند
new_df_1 =df1.dropna(axis=1)
new_df_1
```

### مهم
##### حذف سطر یا ستون ها که مقادیر مشخصی داده نال دارند (مقادیر از دست رفته شان زیاد نباشد)
+   گذاشتن سقف

```python
# هر ردیفی که حداقب دو مقدار سالم داشته باشد باقی میماند
new_df_2 = df1.dropna(axis=0 ,thresh=2)
new_df_2
```

### جایگزینی مقادبر از دست رفته
+ جایگزینی با صفر
+   با میانگین
+   با مقادیر قبلی

```python
new_df_3=df1.fillna(0)
new_df_3
```

```python
## by mean
# numeric_only=True
new_df_3=df1.fillna(df1.mean(numeric_only=True))
new_df_3
```

```python
## by mean rolling 4
new_df_3=df1['A'].fillna(df1['A'].rolling(3).mean())
new_df_3
```

```python
## با مقادبر قبلی پر می شود
new_df_4=df1.ffill()
new_df_4
```

```python
## با مقادبر بعدی پر می شود
new_df_5=df1.bfill()
new_df_5
```

```python
## بامیانگین  مقادبر قبلی وبعدی پر می شود
df2=df1.copy()
df2['A']= (df2['A'].bfill() + df2['B'].ffill())/2
df2
```
