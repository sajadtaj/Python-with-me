## While
+   تا زمانی که شرط برقرا باشد کد های درون بلوک اجرا میگردد
+   مراقب باشید در حلقه بی نهایت نیافتد

```python
countir = 5
while countir > 0:
    print(countir)
    countir -=1
```

## حلقه بی نهایت

```python
i = 0
x = 5
while i == 0 :
    if x==1 :
        print(i)
        i -=1
    else:
        print(x)

```
